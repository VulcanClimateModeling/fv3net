### Differentiate data

[K/m]

```python
dtemp_dx = data_3d_re.temp.differentiate('lon', 2)
dtemp_dy = data_3d_re.temp.differentiate('lat', 2)
```

### Convert to metric units

From K/0.9375deg to K/m: K/m = K/0.9375deg * 0.9375deg/m

Same as dividing by dlat, dlon

```python
dtemp_dx.shape, np.array(dlon_m).shape
```

```python
dtemp_dx = dtemp_dx / np.array(dlon_m)[None, None, :, None]
```

```python
dtemp_dy = dtemp_dy / np.array(dlat_m)[None, None, :, None]
```

## Compute advection term

```python
adv = data_3d_re.u * dtemp_dx + data_3d_re.v * dtemp_dy
```

```python
fig, axs = plt.subplots(3, 2, figsize=(10, 10))
data_3d_re.temp.sel(lat=slice(-40, -20), lon=slice(120, 140)).isel(time=0, pfull=20).plot(ax=axs[0, 0])
dtemp_dx.sel(lat=slice(-40, -20), lon=slice(120, 140)).isel(time=0, pfull=20).plot(ax=axs[1, 1])
data_3d_re.u.sel(lat=slice(-40, -20), lon=slice(120, 140)).isel(time=0, pfull=20).plot(ax=axs[1, 0])
adv.sel(lat=slice(-40, -20), lon=slice(120, 140)).isel(time=0, pfull=20).plot(ax=axs[0, 1])
dtemp_dy.sel(lat=slice(-40, -20), lon=slice(120, 140)).isel(time=0, pfull=20).plot(ax=axs[2, 1])
data_3d_re.v.sel(lat=slice(-40, -20), lon=slice(120, 140)).isel(time=0, pfull=20).plot(ax=axs[2, 0])
```

## Q1

```python
q1 = dtemp_dt + adv
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))
dtemp_dt.sel(lat=slice(-40, -20), lon=slice(120, 140)).isel(time=0, pfull=20).plot(ax=ax1)
adv.sel(lat=slice(-40, -20), lon=slice(120, 140)).isel(time=0, pfull=20).plot(ax=ax2)
q1.sel(lat=slice(-40, -20), lon=slice(120, 140)).isel(time=0, pfull=20).plot(ax=ax3)
plt.tight_layout()
```
